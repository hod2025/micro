```python
#select single row by loc[row index]
row_1 = df.loc[1]
row_1
```

```python
#Select Multiple Rows use slicing with .loc[start and end] :
rows_1_2 = df.loc[1:5]
rows_1_2
```

```python
#Select Rows Based on a Condition
filtered_rows = df[df['rollno'] > 24]
filtered_rows
```

```python
#Combine row and column selection using .loc[]
subset = df.loc[1:2, ['rollno', 'marks']]
subset
```

```python
#filter data using Query
filtered_data = df.query('rollno > 20 and marks < 60')
filtered_data
```
