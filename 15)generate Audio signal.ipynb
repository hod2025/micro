```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import write
output_file_path = 'file_example_WAV_1MG.wav'
```

```python
min_val=-4*np.pi
max_val=4*np.pi
```

```python
time = np.linspace(min_val,max_val, 4*44100)
audio_signal =np.sin(2 * np.pi * 784* time)
```

```python
noise=0.5*np.random.rand(4*44100)
audio_signal+=noise
audio_signal
```

```python
scaling_factor=np.power(2,15)-1
signal_normalization=audio_signal/np.max(np.abs(audio_signal))
signal_scaled=np.int16(signal_normalization*scaling_factor)
```

```python
write(output_file_path,44100,signal_scaled)
```

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(audio_signal, sr=44100)
plt.title('Generated Audio Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
