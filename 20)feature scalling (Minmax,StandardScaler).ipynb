```python
import pandas as pd
import numpy as np
```

```python
data='diabetes.csv'
names=['A','B','C','D','E','F','G','H','I']
a=pd.read_csv(data,names=names)
a
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
mx=MinMaxScaler(feature_range=(0,1))
rs=mx.fit_transform(a)
rs
```

```python
sx=StandardScaler().fit(a)
rx=sx.transform(a)
rx
```
